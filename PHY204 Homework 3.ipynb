```python
# constant 
c = 3 * 10**8 
e = 1.6 * 10**(-19) 
B = 4 * 10**(-5) 
m = 9 * 10**(-31) 
N = 5 * 10**10 
epsilon = 10**(-9) / (36 * np.pi) 

omega_c = e * B / m 
omega_p = ( (N * e**2) / (m * epsilon) )**(1/2) 
```

```python
def k_plus_re(omega): 
    return (omega**2 - (omega_p**2 * omega) / (omega - omega_c))**(1/2) / c 

def k_plus_im(omega): 
    return ((omega_p**2 * omega) / (omega - omega_c) - omega**2)**(1/2) / c 

def k_minus_re(omega): 
    return (omega**2 - (omega_p**2 * omega) / (omega + omega_c))**(1/2) / c 

def k_minus_im(omega): 
    return ((omega_p**2 * omega) / (omega + omega_c) - omega**2)**(1/2) / c 
```

```python
n = 100 
w = omega_c / n 

W = [] 
K_plus_re = [] 
K_plus_im = [] 
K_minus_re = [] 
K_minus_im = [] 

for i in range(1, 5*n): 
    omega = i * w 
    if (omega != omega_c): 
        kpr = k_plus_re(omega) 
        kpi = k_plus_im(omega) 
        kmr = k_minus_re(omega) 
        kmi = k_minus_im(omega) 
    W.append(omega) 
    K_plus_re.append(kpr) 
    K_plus_im.append(kpi) 
    K_minus_re.append(kmr) 
    K_minus_im.append(kmi) 
```

```python
plt.plot(W, K_plus_re, lw=2, color='red', label='real part of k-plus') 
plt.plot(W, K_plus_im, lw=2, color='blue', label='imaginary part of k-plus') 
plt.title('k-plus') 
plt.xlabel('Omega')
plt.ylabel('k') 
plt.legend() 
```
